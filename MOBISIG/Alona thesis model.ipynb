```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import sample
import os
from scipy.stats import zscore
import dtw
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

# Functions

### load signature from the file to data frame

```python
def load_gen_sign(user_num, sign_num):
    user = "USER" + str(user_num)
    sign = "SIGN_GEN_{}_{}_{}".format(user,user,sign_num)
    return pd.read_csv(user + "/" + sign + ".csv")

def load_for_sign(for_user_num, sign_num):
    for_user = "USER" + str(for_user_num)
    for file in os.listdir(for_user):
        if file.endswith(str(sign_num) + ".csv") and file.startswith("SIGN_FOR"): #find the sign- that row needed because we don't kmow who is the user that fake the sign
            sign = file
            break
    return pd.read_csv(str(for_user) + '/' + sign)

#load_for_sign(1, 1)
```

### randomize split the users for train and test and for each user select R set

```python
test_percent = 0.1
user_ammount = 83
test_people_number = int(test_percent * user_ammount)

def select_test(people_number):
    population = [i for i in range(1,84)]
    test_index = sample(population, people_number) # select people for the test set
    train_index = [item for item in population if item not in test_index] #select all the people that not in the test set
    return train_index, test_index

train_ind, test_ind = select_test(test_people_number)
```

```python
def select_R_set(set_size):
    signatures = [i for i in range(1,46)]
    R_ind = sample(signatures, set_size) # select signatures for the reference set
    G_ind = [item for item in signatures if item not in R_ind] #select all the gen signetures that not in the reference set
    return R_ind, G_ind
R_ind, G_ind = select_R_set(5)
```

# save the data from the files to sets

```python
# create train_R
train_R_list = []
train_list = []
for u in train_ind:
    R_u = []
    sign_list = []
    R_ind, G_ind = select_R_set(5)
    for sign in R_ind:
        R_u.append(load_gen_sign(u, sign))
    train_R_list.append(R_u)
    for sign in G_ind: # load gen signatures for the user
        sign_list.append(('G',load_gen_sign(u,sign)))
    for sign in range(1,21): #load forged signatures fir the user
        sign_list.append(('F', load_for_sign(u, sign)))
    train_list.append(sign_list)
    
print('size of train_R_list: {} users'.format(len(train_R_list)))
print('size of R set for each user is {} signatures: '.format(len(train_R_list[0])))
print('size of train_list: {} users'.format(len(train_list)))
print('size of G+F set for each user is {} signatures '.format(len(train_list[0])))
```

```python
# create test_R
test_R_list = []
test_list = []
for u in test_ind:
    R_u = []
    sign_list = []
    R_ind, G_ind = select_R_set(5)
    for sign in R_ind:
        R_u.append(load_gen_sign(u, sign))
    test_R_list.append(R_u)
    for sign in G_ind: # load gen signatures for the user
        sign_list.append(('G',load_gen_sign(u,sign)))
    for sign in range(1,21): #load forged signatures fir the user
        sign_list.append(('F', load_for_sign(u, sign)))
    test_list.append(sign_list)
    
print('size of test_R_list: {} users'.format(len(test_R_list)))
print('size of R set for each user is {} signatures: '.format(len(test_R_list[0])))
print('size of test_list: {} users'.format(len(train_list)))
print('size of G+F set for each user is {} signatures '.format(len(test_list[0])))
```

# Normalize the data

```python
# for the train set

# normalize the R sets
for u in train_R_list:
    for sign in u:
        for dim in list(sign):
            if dim not in ['timestamp', 'gyrox', 'gyroy', 'gyroz']:
                sign[dim] = zscore(sign[dim]) #sign[dim].apply(zscore)

# normalize the G+F sets
for u in train_list:
    for sign in u:
        for dim in list(sign[1]):
            if dim not in ['timestamp', 'gyrox', 'gyroy', 'gyroz']:
                sign[1][dim] = zscore(sign[1][dim]) #sign[dim].apply(zscore)
                
```

```python
df = pd.DataFrame(columns=df_col)
i = 0
for u in range(len(train_list)): #for all the users in the train set
    for sign in range(len(train_list[u])): # for all the signatures of the user
        sign_score_list = [] # list to save the scores vector of the sign
        for dim in list(train_list[u][sign][1])[:10]: # for all the first 10 dim' (without the gyro)
            dim_score_list = [] # list to temporaly save the score for each R signature, from here we will the the min after
            for r in range(len(train_R_list[u])): # for all the R signatures of the user
                x = np.array(train_list[u][sign][1][dim])
                y = np.array(train_R_list[u][r][dim])
                dist, path = fastdtw(x, y, dist = euclidean)
                dim_score_list.append(dist)
            sign_score_list.append(min(dim_score_list))
        sign_score_list.append(train_list[u][sign][0])
        print(sign_score_list)
        df.loc[i] = sign_score_list
        print('row number {} completed'.format(i))
        i +=1
df.to_csv('5_r_75_p_19.3.csv')
```

```python
train = pd.read_csv('5_r_75_p_19.3.csv', index_col= 0 )
test = pd.read_csv('test_r_75_p_19.3.csv', index_col= 0)

x_train = train.drop('label', axis = 1)
y_train = train['label'] == 'F'
x_test = test.drop('label', axis = 1)
y_test = test['label'] == 'F'
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty="l1", #What is our loss function? l2 stands for quad_distances, l1 stands for abs distance.
                             dual=False, #How to define our objective function, as the primal or the dual?
                             tol=0.0001, #Stopping critiria
                             C=1, #Inverse regularization parameter, large C means less regularization.
                             fit_intercept=True, #As it sounds, do we want to learn an intercept?
                             intercept_scaling=1, #What is the initialization of the intercept parameter
                             class_weight=None, #How to treat different classes?
                             random_state=123, #The seed
                             solver="liblinear", #What is the algorithm used to solve the problem?
                             max_iter=10000, #For some optimization algorithms, the algorithm solves it with iterations.
                             verbose=2, 
                             warm_start=False, 
                             n_jobs=1)
lr.fit(x_train,y_train)
```

```python
print(lr.score(x_test, y_test))
probs = lr.predict_proba(x_test)[:,1]
fpr, tpr, threshold  = roc_curve(y_test, probs, pos_label=1)
auc = roc_auc_score(y_test, probs, average=None)
plot_ROC(fpr,tpr,auc)
```
